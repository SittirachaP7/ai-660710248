```python
df = pd.read_csv('student.csv')  # ถ้าไฟล์อยู่ในโฟลเดอร์เดียวกัน
df
```

```python
missing_counts = df.isnull().sum()
print("\n❌ ค่าหายในแต่ละคอลัมน์:")
for col, count in missing_counts[missing_counts > 0].items():
    percentage = (count / len(df)) * 100
    print(f"   - {col}: {count} ค่า ({percentage:.1f}%)")
```

```python
df_clean = df.dropna()
print(f"\nผลการลบ - ก่อนลบ {len(df)} แถว, หลังลบ {len(df_clean)} แถว")
df_clean
```

```python
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
mean_cm = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'].fillna(mean_cm, inplace=True)
mean_kg = df['น้ำหนัก_kg'].mean()
df['น้ำหนัก_kg'].fillna(mean_kg, inplace=True)
mean_hr = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hr, inplace=True)
```

```python
df['ชื่อ'].fillna(df['ชื่อ'].mode()[0], inplace=True)
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวกรรม', 'Engineering','วิศวะ','วิศวฯ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

```python
def standardize_faculty(sex):
    if sex in ['ชาย', 'Male', 'M','ช']:
        return 'ชาย'
    elif sex in ['หญิง', 'Female', 'F','ญ']:
        return 'หญิง'
    else:
        return sex
```

```python
df['เพศ_มาตรฐาน'] = df['เพศ'].apply(standardize_faculty)
```

```python
print("ข้อมูลต้นฉบับ")
print(df)
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```
